# **Randomness is the foundation upon which the Central Limit Theorem (CLT) stands.** Every time we take a sample from a population, randomness determines which elements are included  and this variability causes each sample mean to differ. The CLT reveals a powerful insight: despite this randomness, when we take many random samples and compute their means, the distribution of those means tends to form a predictable, normal (bell-shaped) curve, regardless of the shape of the original population.

image.png

**The Central Limit Theorem (CLT)** explains how we can find patterns in randomness. When we take a random sample from a group (like choosing 30 people at random from a city), the results can vary  each sample will likely give us a different average (like different average heights or incomes). This happens because of **randomness**. But the amazing part is: if we take lots of these random samples and look at all the sample averages, the shape of their distribution will form a bell curve  even if the original data wasn't bell-shaped. This means randomness at the sample level turns into something predictable when repeated many times.
For example, imagine you look at the daily returns of a cryptocurrency, which can go up or down a lot  it's very random and unpredictable. If you randomly pick 30 days and calculate the average return, you'll get different results each time. But if you repeat this many times and plot all the average returns, you'll see a smooth, normal-looking curve. Thats the Central Limit Theorem in action  it shows how repeating random samples can give us reliable patterns we can use to make predictions or build confidence intervals.

## **Randomness**

A random process is one where individual outcomes cannot be determined with certainty, even though there may be a predictable distribution of outcomes over the long run.



**Unpredictability of Individual Outcomes**

In a random process, each outcome is uncertain. For example, when rolling a fair die, you cant predict exactly which number will appear on a single rol


**Defined Probability Distribution**
While individual outcomes are unpredictable, randomness is not chaos. Each possible outcome has a known probability. For the die roll, each of the six numbers has a probability of 1/6


**Reproducible Patterns in the Long Run**
Over many repetitions, random processes tend to show stable patterns. For example, if you roll a fair die thousands of times, each number will occur approximately 1/6 of the time  this is the Law of Large Numbers.


**Fairness and Lack of Bias **
Randomness implies that no outcome is favored over others (unless specified by the probability distribution). In a fair coin toss, both heads and tails are equally likely.

**Example Betting on a Die**
** **conditional** **statement**

Suppose I bet on a roll of a fair die. The rules of the game:

If the die shows 1 spot or 2 spots, I lose a dollar.
If the die shows 3 spots or 4 spots, I neither lose money nor gain money.
If the die shows 5 spots or 6 spots, I gain a dollar.
We will now use conditional statements to define a function one_bet that takes the number of spots on the roll and returns my net gain.




To play the game based on one roll of a die, you can use np.random.choice to generate the number of spots and then use that as the argument to one_bet. Run the cell a few times to see how the output changes.



```python
def one_bet(x):
    """Returns my net gain if the die shows x spots"""
    if x <= 2:
        return -1
    elif x <= 4:
        return 0
    elif x <= 6:
        return 1


one_bet(np.random.choice(np.arange(1, 7)))
```

**The Central Limit Theorem**


The Central Limit Theorem says that the probability distribution of the sum or average of a large random sample drawn with replacement will be roughly normal, regardless of the distribution of the population from which the sample is drawn.




If you take many random samples of the same size from any population, and calculate the sample mean each time, the distribution of those means will be approximately normal (bell-shaped)  e**ven if the original population is not normal**  as long as the sample size is large enough (***typically n  30***).

# **Pizza Delivery Times**


Lets say a pizza delivery company has wildly variable delivery times (e.g., from 10 to 60 minutes). The distribution is not normal  it is skewed, maybe because there are more short deliveries than long ones.

But, if you randomly take samples of 30 deliveries many times and calculate the average time for each sample, the distribution of those sample means will look normal.



```python
import numpy as np
import matplotlib.pyplot as plt

# Skewed population (simulate long-tail delivery times)
np.random.seed(42)
population = np.random.exponential(scale=10, size=100000) + 20  # Skewed population

# Function to simulate sampling distribution
def simulate_sampling_distribution(sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    return np.array(sample_means)

# Simulate sampling distribution
sample_size = 30
num_samples = 10000
sample_means = simulate_sampling_distribution(sample_size, num_samples)

# Step 4: Plotting
plt.figure(figsize=(12, 6))

# Plot original population
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='salmon', edgecolor='black')
plt.title("Original Skewed Population")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")

# Plot sampling distribution of the mean
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, color='skyblue', edgecolor='black')
plt.title(f"Sampling Distribution (n={sample_size})")
plt.xlabel("Sample Mean Delivery Time")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

# ***Finance Example: Daily Stock Returns***

Imagine youre a financial analyst studying the daily returns of a volatile stock (like a tech stock). The returns are not normally distributed  they might be skewed or have fat tails. You want to understand:

What does the distribution of average returns over 30 days look like?
This is where the Central Limit Theorem comes in: it tells us that if you repeatedly take 30-day samples and compute the mean return, the distribution of those means becomes normal, even if the individual daily returns are not.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


sns.set(style="whitegrid")

#   a skewed population of daily stock returns
np.random.seed(42)
daily_returns = np.random.exponential(scale=1.5, size=100000) - 1  # range approx: -1% to ~10%

#  sampling distribution of 30-day average returns
def simulate_clt(data, sample_size=30, num_samples=10000):
    sample_means = [np.mean(np.random.choice(data, size=sample_size, replace=True))
                    for _ in range(num_samples)]
    return np.array(sample_means)

sample_means = simulate_clt(daily_returns)

#  Put data into DataFrame for Seaborn
df = pd.DataFrame({
    'Daily Returns': daily_returns,
    'Type': 'Original Daily Returns'
})
df_means = pd.DataFrame({
    'Daily Returns': sample_means,
    'Type': '30-Day Sample Mean Returns'
})
df_all = pd.concat([df, df_means])

# Plot  histogram with by Seaborn
plt.figure(figsize=(14, 6))

sns.histplot(data=df_all, x='Daily Returns', hue='Type', bins=60, kde=True,
             element='step', stat='density', common_norm=False, palette='Set2')

plt.show()

```

Shows how volatility smooths out over time.

Helps justify the use of normal-based models (e.g., Sharpe Ratio, confidence intervals) when working with average returns.

Makes risk analysis (e.g., for 30-day holding periods) more reliable.
